```python
# 5-11, P118 : 
# 'cty'변수와 'hwy'를 이용하여 'total'(통합연비) 파생변수 만들기

df['total']=(df.cty+df.hwy)/2
df.head()
# df['total']=(df[:,'cty']+df[:,'hwy'])/2
# df['total']=(df.loc[:,'cty']+df.loc[:,'hwy'])/2
# df.total=(df.cty+df.hwy)/2 # UserWarning: Pandas doesn't allow columns to be created via a new attribute name 
```

```python
# 5-14, P120 : 
# 'total'(통합연비) 20이상이면 'pass', 아니면 'fail'의 파생변수 'test' 생성하기

import numpy as np
df['test']=np.where(df.total>=20,'pass','fail')
df.head()
```

```python
# 5-15, P121 : 
# 'test' 변수의 빈도표 만들기

df.test.value_counts()
# df['test'].value_counts()
```

```python
# 5-17, P123 : 
# 'test' 막대그래프 X축 이름 수평으로 바꾸기

df.test.value_counts().plot.bar(rot=0)
```

```python
# 5-18, P124 : 
# 'total'(통합연비) 30이상이면 'A', 20~29이면 'B', 20미만이면 'C'의 파생변수 'grade' 생성하기

df['grade']=np.where(df.total>=30, 'A', np.where(df.total>=20,'B','C'))
df.head()
```

```python
# 5-19, P124 : 
# 'grade' 빈도표 만들기

df.grade.value_counts()
```

```python
# 5-20, P125 : 
# 'grade' 막대그래프 그리기

df.grade.value_counts().plot.bar(rot=0)
```

```python
# 5-21, P126 : 
# 'grade' 막대그래프 알파벳 순으로 정렬하기
# 메서드 체이닝 (method chaining)

df.grade.value_counts().sort_index().plot.bar(rot=0)
```

```python
# 5-22, 추가 : 
# 'grade' 막대그래프 알파벳 역순으로 정렬하기

df.grade.value_counts().sort_index(ascending=False).plot.bar(rot=0)
```

```python
# 5-23, P128 : 
# | (vertical bar)를 이용한 파생변수 생성하기
# 'category' 변수의 compact, subcompact, 2seater이면 'small', 아니면 'large'의 파생변수 'size' 생성하기

df['size']=np.where((df.category=='compact')|(df.category=='subcompact')|(df.category=='2seater'),'small','large')
df.head()
```

```python
# 5-24, 추가 : 
# 'size' 변수를 제외하기

df=df.loc[:,'manufacturer':'grade']
df.head()
```

```python
# 5-25, P128 : 
# isin() 함수를 사용하여 파생변수 생성하기
# 'category' 변수의 compact, subcompact, 2seater이면 'small', 아니면 'large'의 파생변수 'size' 생성하기

df['size']=np.where(df.category.isin(['compact','subcompact','2seater']),'small','large')
df.head()
```
